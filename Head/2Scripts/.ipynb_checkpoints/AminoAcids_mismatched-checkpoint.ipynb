```python
# ПО КОЛИЧЕСТВУ АМИНОКИСЛОТ

result_q = pd.DataFrame(columns = ['Unnamed: 0', 'OrganismName', 'GenName', 'GenID', 'TranslatedAminoAcids', 'GenBankAminoAcids'])

for file in dirname:
    try:
        df = pd.read_csv((os.path.join(direct, file)), sep='\t')
        result_q = result_q.append(df.loc[(df['TranslatedAminoAcids'].apply(lambda x: len(x)) != df['GenBankAminoAcids'].apply(lambda x: len(x)))])
    except:
        pass
    
result_q
```

```python
result_q['Q'] = result_q['TranslatedAminoAcids'].apply(lambda x: len(x)) - result_q['GenBankAminoAcids'].apply(lambda x: len(x))
```
